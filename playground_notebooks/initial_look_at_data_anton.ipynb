```python
# show pose data: ekf_ori, tango_ori, tango_pos
fig, ax = plt.subplots(1, 3, figsize=(12, 4))
ax[0].plot(data_dict['pose/tango_pos'][:,0],
            data_dict['pose/tango_pos'][:,1],
            label = 'tango_pos',
            #c=data_dict['pose/tango_pos'][:,2],
        )
ax[0].legend()
ax[0].set_title('pos')

ax[1].plot(data_dict['pose/tango_ori'][:,0],
            data_dict['pose/tango_ori'][:,1],
            #c=data_dict['pose/tango_ori'][:,2],
        )
ax[1].set_title('tango_ori')
ax[2].plot(data_dict['pose/ekf_ori'][:,0],
            data_dict['pose/ekf_ori'][:,1],
            #c=data_dict['pose/ekf_ori'][:,2],
        )
ax[2].set_title('ekf_ori')
plt.tight_layout()
plt.show()
```
